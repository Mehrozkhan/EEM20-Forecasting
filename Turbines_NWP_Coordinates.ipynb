```
def shortest_distance(df_1, df_2):
    # Haversine Formula to find distance between two coordinates
    R = 6373.0
    lat1 = radians(df_1[1])
    lon1 = radians(df_1[0])
    lat2 = radians(df_2[1])
    lon2 = radians(df_2[0])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    s.append(distance)
    

m= []

for df_1 in turbine.new_col:
    s=[]
    for values in geo_coord.new_col:
        shortest_distance(df_1,values)
    m.append(s.index(min(s)))
coor= [geo_coord.new_col[co] for co in m] #coordinates of NWP data to be used against each turbine
```

```
coordinates= coor
coordinates
```

```
longitude, latitude = [list(tup) for tup in zip(*coordinates)]
```

```
coordinates_NWP = pd.DataFrame(data=list(zip(longitude,latitude)),columns=['longitude', 'latitude'])
```

```
coordinates_NWP.to_csv('Turbine_NWP_coordinates.csv')
```
